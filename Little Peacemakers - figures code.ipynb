```python
def HD_generations(xH_0, xD_0, tH, tD, V, b, c, N):

    xH = [xH_0]
    xD = [xD_0]

    Ehh = 1 + 0.5*b - 0.5*c
    Ehd = 1 + b
    Edh = 1
    Edd = 1 + 0.5*b
    I_d_d = Edd
    I_h_h = Ehh
    I_h_d = Ehd*(1-tD) + Edh*tH
    I_d_h = Edh*(1-tH) + Ehd*tD
    
    for i in range(0,N):
        xHW_int = xH[i]*xH[i]*I_h_h + xH[i]*xD[i]*I_h_d
        xDW_int = xD[i]*xD[i]*I_d_d + xD[i]*xH[i]*I_d_h
        W = xHW_int + xDW_int
        xH_int = xHW_int/W
        xD_int = xDW_int/W
        xH_next = xH_int*(V) + (1-V)*((xH[i]*xH[i]) + (xH[i]*xD[i]*(1-tD+tH)))
        xD_next = 1-xH_next
        xH.append(xH_next)
        xD.append(xD_next) 
        
    return xH  


def mean_fitness_func_T(tD_tH_ratio, b_c_ratio, c, N):
    mean_W = []
    b = b_c_ratio*c
    gT_axis = np.linspace(0,0.5,N) 
    
    for i in range(0,N):
        T = gT_axis[i]
        tH = T
        tD = tD_tH_ratio*tH
        xH = stable_state_perfect_VT(tH, tD, b, c)
        mean_w_m = 1 + 0.5*b - 0.5*c*xH**2
        xH = stable_state_perfect_VT(0, 0, b, c)
        mean_w_g = 1 + 0.5*b - 0.5*c*xH**2
        mean_W.append(mean_w_m-mean_w_g)
        
    return mean_W


def find_min_max_fitness(tD_tH_ratio, c, N):
    fitness_min = mean_fitness_func_T(tD_tH_ratio, 0.0001, c, N)
    fitness_max = mean_fitness_func_T(tD_tH_ratio, 0.0001, c, N)
    b_c_ratio_axis = np.linspace(0.0001, 1.2, N)    
    
    for i in range(0,N):
        b_c_ratio = b_c_ratio_axis[i]
        cur_mean_fitness = mean_fitness_func_T(tD_tH_ratio, b_c_ratio, c, N)
        for j in range(0,N):
            fitness_min[j] = min(fitness_min[j], cur_mean_fitness[j])
            fitness_max[j] = max(fitness_max[j], cur_mean_fitness[j])
        
    return fitness_min, fitness_max



def plot_Fig3(xH_genes_arr, xH_09_arr, xH_095_arr, xH_1_arr, xH_105_arr, fitness_min_arr, fitness_max_arr):

    fig3,axs = plt.subplots(1,3 ,figsize=(13,3.5), gridspec_kw={'width_ratios':[1,0.2,1]})

    #Fig 3.a
    # axs[0].plot(xH_genes_arr[0][0], linewidth=3,c='k')
    # axs[0].plot(xH_genes_arr[1][0], linewidth=3,c='k',linestyle="dashed" )
    # axs[0].plot(xH_genes_arr[2][0], linewidth=3,c='k',linestyle="dotted" )
    
    axs[0].plot(xH_09_arr[0], linewidth=3,c='k')
    axs[0].plot(xH_09_arr[1], linewidth=3,c='k',linestyle="dashed" )
    axs[0].plot(xH_09_arr[2], linewidth=3,c='k',linestyle="dotted" )
    
    axs[0].plot(xH_095_arr[0], linewidth=3,c='#ad178d')
    axs[0].plot(xH_095_arr[1], linewidth=3,c='#ad178d',linestyle="dashed" )
    axs[0].plot(xH_095_arr[2], linewidth=3,c='#ad178d', linestyle="dotted" )
    
    axs[0].plot(xH_1_arr[0], linewidth=3,c='#f37651')
    axs[0].plot(xH_1_arr[1], linewidth=3,c='#f37651',linestyle="dashed" )
    axs[0].plot(xH_1_arr[2], linewidth=3,c='#f37651', linestyle="dotted" )
    
    axs[0].plot(xH_105_arr[0], linewidth=3,c='#C9C9C9')
    axs[0].plot(xH_105_arr[1], linewidth=3,c='#C9C9C9',linestyle="dashed" )
    axs[0].plot(xH_105_arr[2], linewidth=3,c='#C9C9C9', linestyle="dotted" )
    
    axs[0].plot(xH_genes_arr[0], linewidth=3,c='C0')

    axs[0].set_xscale("symlog")
    axs[0].legend([r'$VT=1$' ,r'$VT=0.5$',r'$VT=0.1$'], fontsize=12,ncol=1, fancybox=True, shadow=True)
    axs[0].set_xlabel(r'$Generations$', fontsize=18)
    axs[0].set_ylabel(r'$Frequency$' +" "+ r'$of$' +" "+ r'$Hawk$', fontsize=18)
    axs[0].tick_params(labelsize=16)
    axs[0].set_yticks([0,0.2,0.4,0.6,0.8,1])
    axs[0].set_ylim(-0.05, 1.05)
    axs[0].set_xlim(1, 10000)

    axs[0].axhline(y=-0.05, color='black',linewidth=4, alpha=0.7)
    axs[0].axhline(y=1.05, color='black',linewidth=4, alpha=0.7)
    axs[0].axvline(x=1, color='black',linewidth=4, alpha=0.7)
    axs[0].axvline(x=10000, color='black',linewidth=4, alpha=0.7)


    axs[1].axis('off')
    
    #Fig 3.b
    T_axis=np.linspace(0,0.5,1000)
    
    axs[2].plot(T_axis[0:-1],fitness_min_arr[3][0:-1], linewidth=3,c="#C9C9C9", alpha=0.8)
    axs[2].plot(T_axis[0:-1],fitness_min_arr[2][0:-1], linewidth=3,c="#f37651", alpha=0.8)
    axs[2].plot(T_axis[0:-1],fitness_min_arr[1][0:-1], linewidth=3,c="#ad178d", alpha=0.8)
    axs[2].plot(T_axis[0:-1],fitness_min_arr[0][0:-1], linewidth=3,c="k", alpha=0.8)
    
    axs[2].plot(T_axis[0:-1],fitness_max_arr[0][0:-1], linewidth=3,c="k", alpha=0.8)
    axs[2].plot(T_axis[0:-1],fitness_max_arr[1][0:-1], linewidth=3,c="#ad178d", alpha=0.8)
    axs[2].plot(T_axis[0:-1],fitness_max_arr[3][0:-1], linewidth=3,c="#C9C9C9", alpha=0.8)
    axs[2].plot(T_axis[0:-1],fitness_max_arr[2][0:-1], linewidth=3,c="#f37651", alpha=0.8)    
    
    axs[2].legend([r'$T_D$'+r'$_H$'+ "={0}".format(1.05)+ r'$\cdot T_H$'+r'$_D$',r'$T_D$'+r'$_H$'+ "=" + r'$T_H$'+r'$_D$',r'$T_D$'+r'$_H$'+ "={0}".format(0.95)+ r'$\cdot T_H$'+r'$_D$'  ,r'$T_D$'+r'$_H$'+ "={0}".format(0.9)+ r'$\cdot T_H$'+r'$_D$'] ,ncol=1,fontsize=14,loc='center left', bbox_to_anchor=(1.1, 0.7), fancybox=True, shadow=True)
    axs[2].set_xlabel(r'$T_H$'+r'$_D$' , fontsize=18)
    axs[2].set_ylabel(r'$\Delta$'+" "+r'$ Mean$'+" "+r'$Fitness$' , fontsize=18)
    
    axs[2].tick_params(labelsize=16)
    axs[2].set_ylim(-0.055, 0.055)
    axs[2].set_xlim(0, 0.5)
    axs[2].set_yticks([-0.05,-0.025,0,0.025,0.05])
    
    axs[2].axhline(y=-0.055, color='black',linewidth=4, alpha=0.7)
    axs[2].axhline(y=0.055, color='black',linewidth=4, alpha=0.7)
    axs[2].axvline(x=0, color='black',linewidth=4, alpha=0.7)
    axs[2].axvline(x=0.5, color='black',linewidth=4, alpha=0.7)
    
    axs[2].fill_between(T_axis[0:-1],fitness_min_arr[2][0:-1],fitness_max_arr[2][0:-1], color='#f37651', alpha=0.6 ,interpolate=True ,hatch='..',edgecolor="#f37651")
    axs[2].fill_between(T_axis[0:-1], fitness_min_arr[1][0:-1],fitness_max_arr[1][0:-1], color='#ad178d', alpha=0.4,interpolate=True)
    axs[2].fill_between(T_axis[0:-1],fitness_min_arr[3][0:-1],fitness_max_arr[3][0:-1], color='#C9C9C9', alpha=0.5,interpolate=True)
    axs[2].fill_between(T_axis[0:-1], fitness_min_arr[0][0:-1],fitness_max_arr[0][0:-1], color='k', alpha=0.6, interpolate=True)
    
    # a b
    axs[0].text(0.35, 1.15, r'$a$', fontsize=26)
    axs[2].text(-0.06, 1.25*0.05, r'$b$', fontsize=26)

    axs[0].text(165, 0.63, r'$Genetic~case~(T=0)$', c='C0', fontsize=12)

    
    fig3.set_dpi(330)
    fig3.show()
```

```python
def A_B_microbes_sim_pA_T(bc_ratio, N):
    T_max=0.5
    pA_axis = np.linspace(0,1,N) # y-axis
    Ta_g_axis = np.linspace(0,T_max,N) # x-axis
    freq_mat = np.zeros((N,N))
    H_mat = np.zeros((N,N))
    
    for i in range(0,N):
        T = Ta_g_axis[i]
        for k in range(0,N):
            pA = pA_axis[k]
            P_microbes=(bc_ratio)*(1-2*T)
            if pA>=bc_ratio:    
                freq_mat[k,i]=0
            elif pA<=P_microbes:
                freq_mat[k,i]=((bc_ratio)-P_microbes)/((bc_ratio)-pA)
            else:
                freq_mat[k,i]=1
            H_mat[k,i]=freq_mat[k,i]*pA +  (1-freq_mat[k,i])*(bc_ratio)
            
    return freq_mat,H_mat 


def plot_Fig4_ab(freq_mat_1, freq_mat_2, bc_1, bc_2, N):
    T_max=0.5
    bc_max=1
    txtX, txtY = r'$T$', r'$P_A(H)$' 
    xNticks, yNticks = 6, 6
    dec_x, dec_y = 2, 2    
    gT_axis = np.around(np.linspace(0,T_max,xNticks),decimals=dec_x) # x-axis
    bc_axis = np.around(np.linspace(0,bc_max,yNticks),decimals=dec_y) # y-axis
    Y_lbl, X_lbl = bc_axis, gT_axis
    
    fig4a,axs = plt.subplots(1,4, gridspec_kw={'width_ratios':[1,0.001,1,0.1]} ,figsize=(8.8,4))

    # cbar_ax = fig4a.add_axes([0.94, 0.22, 0.035, 0.6])
    cbar_ax = fig4a.add_axes([0.94, 0.19, 0.035, 0.6])

    cbar_ax.set_title(r'$Proportion$'+'\n' + r'$of$' + ' ' + r'$A$', fontsize=18)
    cbar_ax.tick_params(labelsize=14)
    cmap=sns.color_palette("GnBu_r", as_cmap=True)
    
    axs[0] = sns.heatmap(freq_mat_1,xticklabels=N//(xNticks-1), yticklabels=N//(yNticks-1), ax=axs[0], vmin=0, vmax=1,cmap=cmap, cbar=False)
    axs[0].set_xticklabels(X_lbl,fontsize=12)
    axs[0].set_yticklabels(Y_lbl,rotation=0,fontsize=12)
    axs[0].set_xlabel(txtX, size=20)
    axs[0].set_ylabel(txtY, size=20)
    
    cs=axs[0].contour(freq_mat_1,[0.2,0.4,0.6,0.8], colors='k')
    manual_locations = [(N*0.08/0.5,N*0.1),(N*0.16/0.5,N*0.1),(N*0.25/0.5,N*0.1),(N*0.34/0.5,N*0.1)]
    clbl=axs[0].clabel(cs,inline=True,fontsize=9,inline_spacing=5, manual=manual_locations)
    for l in clbl:
        l.set_rotation(0)
    cs=axs[0].contour(freq_mat_1,[0.998], colors='k', linewidths=3)
    manual_locations = [(N*0.43/0.5,N*0.1)]
    clbl=axs[0].clabel(cs,inline=True,fontsize=10,inline_spacing=4, manual=manual_locations, fmt="1.0")
    for l in clbl:
        l.set_rotation(0)
    x1= np.linspace(0,T_max,N)
    y1=(bc_1)*(1-2*x1) 
    y3=np.ones(N)
    axs[0].axhline(y=N*bc_1, color='k',linewidth=3)
    axs[0].fill_between(x1*N/T_max,bc_1*N,y3*N ,color='steelblue',hatch=".",edgecolor="navy", alpha=0.5)
    axs[0].fill_between(x1*N/T_max,y1*N,bc_1*N ,color='whitesmoke')
    axs[0].set_title(r'$P_B=b/c=0.7$', size=20)

    axs[2] = sns.heatmap(freq_mat_2,xticklabels=N//(xNticks-1), yticklabels=N//(yNticks-1), ax=axs[2],cbar_ax=cbar_ax, vmin=0, vmax=1,cmap=cmap)
    axs[2].set_xticklabels(X_lbl,fontsize=12)
    axs[2].set_yticklabels(Y_lbl,rotation=0,fontsize=12)
    axs[2].set_xlabel(txtX, size=20)
    axs[2].set_ylabel(txtY, size=20)
    cs=axs[2].contour(freq_mat_2,[0.2,0.4,0.6,0.8], colors='k')
    manual_locations = [(N*0.08/0.5,N*0.1),(N*0.16/0.5,N*0.1),(N*0.25/0.5,N*0.1),(N*0.34/0.5,N*0.1)]
    clbl=axs[2].clabel(cs,inline=True,fontsize=9,inline_spacing=5, manual=manual_locations)
    for l in clbl:
        l.set_rotation(0)
    cs=axs[2].contour(freq_mat_2,[0.998], colors='k', linewidths=3)
    manual_locations = [(N*0.43/0.5,N*0.1)]
    clbl=axs[2].clabel(cs,inline=True,fontsize=10,inline_spacing=4, manual=manual_locations, fmt="1.0")
    for l in clbl:
        l.set_rotation(0)
    x1= np.linspace(0,T_max,N)
    y1=(bc_2)*(1-2*x1) 
    y3=np.ones(N)
    axs[2].axhline(y=N*bc_2, color='k',linewidth=3)
    axs[2].fill_between(x1*N/T_max,bc_2*N,y3*N ,color='steelblue',hatch=".",edgecolor="navy", alpha=0.5)
    axs[2].fill_between(x1*N/T_max,y1*N,bc_2*N ,color='whitesmoke')
    axs[2].set_title(r'$P_B=b/c=0.9$', size=20)

    for spine in cbar_ax.spines.values():
        spine.set(visible=True, lw=2, edgecolor="black")
    
    for ind in [0,2]:
        axs[ind].axhline(y=0, color='black',linewidth=4, alpha=0.7)
        axs[ind].axhline(y=N, color='black',linewidth=4, alpha=0.7)
        axs[ind].axvline(x=0, color='black',linewidth=4, alpha=0.7)
        axs[ind].axvline(x=N, color='black',linewidth=4, alpha=0.7)
        axs[ind].invert_yaxis()

    axs[1].axis('off')
    axs[3].axis('off')

    
    axs[0].text(-0.15*N, 1.15*N, r'$a$', fontsize=26)
    axs[2].text(-0.15*N, 1.15*N, r'$b$', fontsize=26)

    axs[0].text(0.62*N, 0.92*N, r'$B~fixation$', c='white', fontsize=12)
    axs[0].text(0.62*N, 0.6*N, r'$A~fixation$', c='k', fontsize=12)
    axs[2].text(0.62*N, 0.92*N, r'$B~fixation$', c='white', fontsize=12)
    axs[2].text(0.62*N, 0.8*N, r'$A~fixation$', c='k', fontsize=12)

    axs[0].text(0.4*N, 0.21*N, r'$b/c \cdot (1-2T)$', c='k', fontsize=12, rotation=-34)
    axs[2].text(0.4*N, 0.3*N, r'$b/c \cdot (1-2T)$', c='k', fontsize=12, rotation=-40)
    
    fig4a.set_dpi(330)
    fig4a.tight_layout()
    plt.show()


def plot_Fig4_cd(freq_mat_1, freq_mat_2, bc_1, bc_2, N):

    T_max=0.5
    bc_max=1
    txtX, txtY = r'$T$', r'$P_A(H)$' 
    xNticks, yNticks = 6, 6
    dec_x, dec_y = 2, 2    
    gT_axis = np.around(np.linspace(0,T_max,xNticks),decimals=dec_x) # x-axis
    bc_axis = np.around(np.linspace(0,bc_max,yNticks),decimals=dec_y) # y-axis
    Y_lbl, X_lbl = bc_axis, gT_axis

    fig4c,axs = plt.subplots(1,4, gridspec_kw={'width_ratios':[1,0.001,1,0.1]} ,figsize=(8.8,4))
    cbar_ax = fig4c.add_axes([0.94, 0.19, 0.035, 0.6])
    
    cbar_ax.set_title(r'$Proportion$'+'\n' + r'$of$' + ' ' + r'$Hawk$', fontsize=18)
    cbar_ax.tick_params(labelsize=14)
    cmap=sns.color_palette("rocket_r", as_cmap=True)
    
    axs[0] = sns.heatmap(freq_mat_1,xticklabels=N//(xNticks-1), yticklabels=N//(yNticks-1), ax=axs[0], vmin=0, vmax=1,cmap=cmap, cbar=False)
    axs[0].set_xticklabels(X_lbl,fontsize=12)
    axs[0].set_yticklabels(Y_lbl,rotation=0,fontsize=12)
    axs[0].set_xlabel(txtX, size=20)
    axs[0].set_ylabel(txtY, size=20)
    cs=axs[0].contour(freq_mat_1,[0.1,0.3,0.5], colors='yellow')
    manual_locations = [(N*0.47/0.5,N*0.1),(N*0.47/0.5,N*0.3),(N*0.47/0.5,N*0.5)]
    clbl=axs[0].clabel(cs,inline=True,fontsize=9,inline_spacing=5, manual=manual_locations)
    for l in clbl:
        l.set_rotation(0)
    cs=axs[0].contour(freq_mat_1,[bc_1-0.001], colors='white',linewidths=3)
    manual_locations = [(N*0.47/0.5,N*0.8)]
    clbl=axs[0].clabel(cs,inline=True,fontsize=10,inline_spacing=5, manual=manual_locations,fmt=str(bc_1))
    for l in clbl:
        l.set_rotation(0)
    axs[0].set_title(r'$P_B=b/c=0.7$', size=20)

    axs[2] = sns.heatmap(freq_mat_2,xticklabels=N//(xNticks-1), yticklabels=N//(yNticks-1), ax=axs[2],cbar_ax=cbar_ax, vmin=0, vmax=1,cmap=cmap)
    axs[2].set_xticklabels(X_lbl,fontsize=12)
    axs[2].set_yticklabels(Y_lbl,rotation=0,fontsize=12)
    axs[2].set_xlabel(txtX, size=20)
    axs[2].set_ylabel(txtY, size=20)
    cs=axs[2].contour(freq_mat_2,[0.1,0.3,0.5,0.7], colors='yellow')
    manual_locations = [(N*0.47/0.5,N*0.1),(N*0.47/0.5,N*0.3),(N*0.47/0.5,N*0.5),(N*0.47/0.5,N*0.7)]
    clbl=axs[2].clabel(cs,inline=True,fontsize=9,inline_spacing=5, manual=manual_locations)
    for l in clbl:
        l.set_rotation(0)
    cs=axs[2].contour(freq_mat_2,[bc_2-0.001], colors='white',linewidths=3)
    manual_locations = [(N*0.47/0.5,N*0.8)]
    clbl=axs[2].clabel(cs,inline=True,fontsize=10,inline_spacing=5, manual=manual_locations,fmt=str(bc_2))
    for l in clbl:
        l.set_rotation(0)
    axs[2].set_title(r'$P_B=b/c=0.9$', size=20)
    x1= np.linspace(0,T_max,N)
    y1=(bc_1)*(1-2*x1) 
    y2=(bc_2)*(1-2*x1)
    axs[0].plot(x1*N/T_max,y1*N, color='k',linewidth=3)
    axs[2].plot(x1*N/T_max,y2*N, color='k',linewidth=3)

    for spine in cbar_ax.spines.values():
        spine.set(visible=True, lw=2, edgecolor="black")
    for ind in [0,2]:
        axs[ind].axhline(y=0, color='black',linewidth=4, alpha=0.7)
        axs[ind].axhline(y=N, color='black',linewidth=4, alpha=0.7)
        axs[ind].axvline(x=0, color='black',linewidth=4, alpha=0.7)
        axs[ind].axvline(x=N, color='black',linewidth=4, alpha=0.7)
        axs[ind].invert_yaxis()
        
    axs[1].axis('off')
    axs[3].axis('off')

    axs[0].text(-0.15*N, 1.15*N, r'$c$', fontsize=26)
    axs[2].text(-0.15*N, 1.15*N, r'$d$', fontsize=26)

    
    fig4c.set_dpi(330)

    fig4c.tight_layout()
    plt.show()
```
